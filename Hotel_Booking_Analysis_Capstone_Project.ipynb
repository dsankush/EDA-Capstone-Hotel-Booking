## Adding some new columns to our dataset as per the need and we will use these columns for our analysis

### As we have 'babies' and 'children' columns in our dataset which have the data showing the availability of childrens and babies with the customers.

### And we also have the 'adults'c column in our dataset which is indicating the number of adults booked the hotel.



* <b> We will take these variables together to get the total number of members arrived in a particular hotel by adding another column to our dataset</b>


1. Adding 'total_children' column to our dataset which contains the children and babies


2. Adding 'total_visitors' column to our dataset which contains the 'total_children' and 'adults' together.

```python
# merging two columns 'children' and 'babies' to 'total_children' column

data['total_children'] = data['children'] + data['babies']

# merging 'total_children' and 'adults' to 'total_visitors' column

data['total_visitors'] = data['total_children'] + data['adults']
```

```python
# Converting the data type of newly created columns from float type to integer type using astype from pandas

data['total_children'] = data['total_children'].astype(int)
data['total_visitors'] = data['total_visitors'].astype(int)
```

## As we have total number of childrens count in a single column now we can drop/delete the 'children' and 'babies' column from our data

```python
# Using drop to drop/delete the columns

data.drop(['children','babies'], axis = 1, inplace = True)
```

Visualization is the act of creating images that are used to represent information. * <b>Lets have some observation from hotel type that which type of hotel has how many bookings</b>

```python
# Total Value count of each type of hotel
count_of_hotel_type = data['hotel'].value_counts()
print(f"Total count of each type hotel:\n\n{count_of_hotel_type}")
```

```python
# Using countplot from seaborn to get the count of bookings for each type of hotel

sns.set(font_scale = 1)
sns.set_palette('gist_ncar')
plt.figure(figsize = (10,5))

plt.subplot(1, 2, 1)

sns.countplot(data['hotel'])

plt.title('Percent of hotel_type bookings')
plt.xlabel('Hotel Type')
plt.ylabel('Number of bookings made')


# Using pie from matplot to get the percent of bookings for each type of hotel

plt.subplot(1, 2, 2)

plt.pie(list(count_of_hotel_type), labels=['City Hotel', 'Resort Hotel'], 
        autopct='%1.01f%%',startangle=90, explode= (0, 0.1))

plt.title('Percent of hotel_type bookings')


plt.show()
```

# Inference:
1. City hotel has bookings above 50000 which is around 61.4% and Resort hotel has less than 35000 which is around 38.6% bookings


2. Which implies that City Hotel has more number of bookings than Resort Hotel

```python
# Total Value count of hotel booking cancellation

print(f"Total count of cancelation and not cancellation :\n\n{data['is_canceled'].value_counts()}")
```

```python
# Using countplot from seaborn to get the count of bookings cancelled or not

# 0 ---> Non Cancelled Booking
# 1 ---> Cancelled Booking

plt.figure(figsize = (10,5))

sns.countplot(data['is_canceled'])

plt.title('Count of hotel booking cancellation')
plt.xlabel('Canceld or Not Canceled')
plt.ylabel('Number of cancellation')
plt.show()
```

. <b>Now Checking number of cancellation done by for each type hotels</b>

```python
# Seperating only that data which has cancellation 
canceled_hotel = data[data['is_canceled'] == 1]

# grouping type of hotels and only booking cancellation using groupby method

canceled_hotel = canceled_hotel.groupby(['hotel']).size().reset_index(name = 'cancellation_count')

print('Groupby type of hotel and booking cancelled data')

canceled_hotel
```

```python
# Ploting bar graph to visualize the booking cancellation for each type of hotel

plt.figure(figsize = (10,5))
sns.barplot(x = canceled_hotel['hotel'], y = canceled_hotel['cancellation_count'])
plt.title('Count of hotel booking cancellation VS for each type of Hotel')
plt.xlabel('Type of Hotel')
plt.ylabel('Number of cancellation')

plt.plot()
```

# Inference:
1. City hotel has more number of cancellation which is about 16000 and Resort Hotel as less cancellation which is less than 8000 which is about almost 50%
